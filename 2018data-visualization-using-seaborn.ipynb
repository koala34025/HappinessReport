# 2018資料分布圖

Importing libraries neccessary for data vizualization and analysis.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Importing the csv file.

```python
df=pd.read_csv('../archive/2018.csv')
df.shape  #checking the shape of the dataset
```

* 156 countries are present in the dataset.

```python
# Defining a method to print the percentage of data values in the plots
def fun_p(colmn,ra=.1):
    plt.figure(figsize=(20,5))
    plt.subplot(1,2,1)
    ax=sns.histplot(data=df, x=colmn, bins=np.arange(df[colmn].min(),df[colmn].max()+ra,ra), kde=True) #plotting a histogram
    total=len(df[colmn]) #taking lenth of each column
    for p in ax.patches:
        percentage='{:.2f}%'.format(100 * p.get_height()/total)  #calculating percentage
        x = p.get_x() + p.get_width() / 2      # width of the plot
        y = p.get_y() + p.get_height()         # height of the plot      
        #ax.annotate(percentage, (x, y), size = 12,ha='center', va='bottom')  # annotate the percentage
    plt.xticks(np.arange(round(df[colmn].min(),1),round(df[colmn].max()+ra,1),ra)) 
    plt.subplot(1,2,2)              
    printx = 0.01
    printy = 0.95
    bx=sns.boxplot(data= df, y = colmn) # plotting boxplot
    # annotate the min, 25%, 50%, 75%, max value on bx
    bx.annotate('0%: {:.2f}'.format(df[colmn].min()), xy=(printx, printy), xycoords='axes fraction')
    bx.annotate('25%: {:.2f}'.format(df[colmn].quantile(.25)), xy=(printx, printy-0.05), xycoords='axes fraction')
    bx.annotate('50%: {:.2f}'.format(df[colmn].quantile(.50)), xy=(printx, printy-0.1), xycoords='axes fraction')
    bx.annotate('75%: {:.2f}'.format(df[colmn].quantile(.75)), xy=(printx, printy-0.15), xycoords='axes fraction')    
    bx.annotate('100%: {:.2f}'.format(df[colmn].max()), xy=(printx, printy-0.2), xycoords='axes fraction')
    # annotate std
    bx.annotate('std: {:.2f}'.format(df[colmn].std()), xy=(printx, printy-0.25), xycoords='axes fraction')
    plt.show()
```

```python
fun_p('Score',ra=.5)
```

* Score is normally distributed
* The IQR ranges between 4.4 and 6.4
* Minumum Score is below 3 and maximum is close to 8.

* Minimum GDP per capita is 0 and maximum is about 15.7 with outliers above 2.
